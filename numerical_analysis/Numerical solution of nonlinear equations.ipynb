### 二分法

```python
#求pi的近似值，利用sin(pi)=0
import numpy as np
from numpy import sin
from numpy import log2
import matplotlib.pyplot as plt

a=3.14
b=3.15
eps=10e-10
res=[]

def f(x):
    return sin(x)

n=int(log2((b-a)/eps))

for i in range(n):
    m=(a+b)/2
    if f(m)*f(a) > 0:
        a=m
    else:
        b=m
    res.append(m)

plt.plot(range(len(res)), res)
plt.xlabel('number of iterations')
plt.ylabel('root')
```

### 牛顿迭代法

```python
import matplotlib.pyplot as plt

#定义参数
n=100 #迭代次数
r=1 #用于都牛顿迭代法修正
x=1 #初值
res=[]

#定义函数
def f(x):
    return x**3

def df(x):
    return 3*x**2

#牛顿迭代
for i in range(n):

    x_next=x-r*f(x)/df(x)
    
    if(x-x_next)< 10e-6:
        break

    #print(x_next)
    res.append(x_next)
    x=x_next


plt.plot(range(len(res)),res)
plt.xlabel('number of iterations')
plt.ylabel('root')
```
