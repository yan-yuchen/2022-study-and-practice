```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np

x=np.array([150,200,250,300,350,400,450])
y=np.array([6450,7450,8450,9450,11450,15450,19450])

# reshape(-1, 1)中的-1是模糊控制的意思 比如人reshape（-1,2）固定2列 多少行不知道
# reshape(-1, 1)将[150 200 250 300 350 400 450]变为[[150] [200] [250] [300] [350] [400] [450]]
# 对于Python中的数组类型ndarray ，(n, )是一维数组，(n, 1)是二维数组，二者形状不一样；
xx=x.reshape(-1, 1)
yy=y.reshape(-1, 1)


model = linear_model.LinearRegression()
model.fit(xx, yy)
```

```python
plt.figure()
plt.scatter(xx,yy)
plt.plot(xx, model.predict(xx), color='red',linewidth=3.0,linestyle="-")
plt.show()
```
